```python
import pandas as pd

df = pd.read_csv('customer_support_tickets.csv')
```

```python
import re

def clean_message(text):
    text = str(text).lower()
    text = re.sub(r'[^a-z0-9\s]', ' ', text)
    text = re.sub(r'\s+', ' ', text).strip()
    return text

df['clean_message'] = df['Ticket Description'].apply(clean_message)
```

```python
df[['Ticket Description', 'clean_message']].head()
```

```python
def classify_issue(text):
    if 'payment' in text or 'card' in text or 'upi' in text:
        return 'PAYMENT'
    elif 'login' in text or 'password' in text or 'otp' in text:
        return 'LOGIN'
    elif 'delivery' in text or 'delay' in text:
        return 'DELIVERY'
    elif 'refund' in text or 'return' in text:
        return 'REFUND'
    elif 'error' in text or 'bug' in text:
        return 'BUG'
    else:
        return 'GENERAL'

df['issue_type'] = df['clean_message'].apply(classify_issue)
```

```python
df['issue_type'].value_counts()
```

```python
def assign_priority(text):
    if 'urgent' in text or 'critical' in text or 'immediately' in text:
        return 'P0'
    elif 'failed' in text or 'important' in text:
        return 'P1'
    elif 'delay' in text or 'slow' in text:
        return 'P2'
    else:
        return 'P3'
df['calculated_priority'] = df['clean_message'].apply(assign_priority)
```

```python
import pandas as pd
from datetime import timedelta

df['created_time'] = pd.to_datetime(
    df['Date of Purchase'],
    dayfirst=True,
    errors='coerce'
)
```

```python
sla_map = {
    'P0': 2,
    'P1': 6,
    'P2': 24,
    'P3': 72
}
```

```python
df['sla_hours'] = df['calculated_priority'].map(sla_map)
```

```python
df['due_time'] = df['created_time'] + df['sla_hours'].apply(
    lambda x: timedelta(hours=x)
)
```

```python
df[['clean_message', 'calculated_priority', 'sla_hours', 'due_time']].head()
```

```python
df.to_csv('final_support_tickets.csv', index=False)
```
